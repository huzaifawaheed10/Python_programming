```python
sq=[1,4,9,16,25,36,49,64,81,100]
i=0
while i<(len(sq)):
    print(sq[i])
    i+=1
```

```python
tup=(1, 4, 9, 16, 25, 36, 49, 64, 81,100)
x=int(input("enter a number please:"))
i=0
while i<len(tup):
    if tup[i]==x:
        print("x found at index",i)
    else:
        print("not found at index",i)
    i+=1
```

```python
num=(1, 4, 9, 16, 25, 36, 49, 64, 81,100)
x=int(input("Enter number please:"))
i=0
for el in num:
    if el==x:
        print("x is found at index",i)
    else:
        print("not found")
    i+=1
```
