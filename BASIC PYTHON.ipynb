```python
numbers=[54,45,8,98,50,10,4,6,2,4]
numbers.sort()
print("number in ascending order is:",numbers)
modified_list=list(sorted(set(numbers)))
print("ascending order modified list :", modified_list)
numbers.sort(reverse=True)
print("numbers in descending order:",numbers)
modified_list=list(sorted(set(numbers),reverse=True))
print("descending order modified list :", modified_list)
```

```python
import random
numbers=[random.randint(50,100) for _ in range(15)]
print("original list :",numbers)
numbers.sort()
print("numbers in ascending order:",numbers)
modified_list_asc=list(sorted(set(numbers)))
print("modified ascending order list :",modified_list_asc)
numbers.sort(reverse=True)
print("numbers in descending order:",numbers)
modified_list_disc=list(sorted(set(numbers),reverse=True))
print("modified descending order list :",modified_list_disc)
```

```python
tup=(1, 4, 9, 16, 25, 36, 49, 64, 81,100)
x=int(input("enter a number please:"))
i=0
while i<len(tup):
    if tup[i]==x:
        print("x found at index",i)
    else:
        print("not found at index",i)
    i+=1
```
