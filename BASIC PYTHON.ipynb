### LISTS AND TUPLES

```python
movies=[]
a=input("enter first movie:")
b=input("enter second movie:")
c=input("enter third movie:")
movies.append(a)
movies.append(b)
movies.append(c)
print(movies)
```

```python
fruit=["Apple","Banana","peach","Grapes","melon"]
print(fruit)
fruit.append("Mango")
print(fruit)
fruit.remove("Banana")
print(fruit)
fruit.reverse()
print(fruit)
fruit.sort()
print(fruit)
fruit.sort(reverse=True)
print(fruit)
fruit.insert(2,"orange")
print(fruit)
```

```python
num=[12,13,14,13,15]
num1=num.copy()
print(num1)
num.reverse()
num2=num.copy()
print(num2)
if num1==num2 :
    print("list is palindrome")
else:
    print("list is not palindrome")
```
