SMART TASK MANAGER

```python
tasks=[]

##a=(input("enter first task"))
#b=(input("enter second task"))
#c=(input("enter third task"))
#d=(input("enter fourth task"))
#e=(input("enter fifth task"))

#tasks.append(a)
#tasks.append(b)
#tasks.append(c)
##asks.append(d)
#tasks.append(e)
tasks = []

for i in range(5):
    task = input("Enter a task: ")
    tasks.append(task)

f=(input("enter which task is completed"))

if f in tasks:
    tasks.remove(f)
    print("task is completed")
else:
    print("task is not  completed")

if not tasks:
    print("list is empty")
else:
    print("remaining tasks:")

for task_name in tasks:
    print(f"*{task_name}")
```

MOVIE RATINGS

```python
ratings=[4, 5, 3, 4, 5, 2]
ratings.append(4.5)

average_rating=round(sum(ratings)/len(ratings),2)
print(f"average rating is {average_rating}")

highest_rating=max(ratings)
lowest_rating=min(ratings)
print(f"highest rating is {highest_rating}")
print(f"lowest rating is {lowest_rating}")

if average_rating>=4:
    print("Excellent")
elif average_rating>=3:
    print("Good")
else:
    print("Needs Improvement")
```

PRODUCT RATING SYSTEM

```python
products=[5, 4, 3, 4, 5, 2]
for i in range(3):
    product=int(input("Enter rating"))
    products.append(product)

average=round(sum(products)/len(products),3)
print(f"average rating is {average}")

good_count = sum(1 for rating in ratings if rating >= 4)
print(f"Number of good ratings (≥ 4): {good_count}")

if good_count>=4:
    print("Popular product")
else:
    print("average product")
```
